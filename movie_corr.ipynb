```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#read data
df = pd.read_csv('./movies.csv')
```

```python
df.info()
```

```python
df.head(5)
```

```python
#find missing values
for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print(f'{col} - {percent_missing}%')
```

```python
df['rating'].value_counts()
```

```python
df['rating'].fillna('R', inplace = True)
```

```python
df['rating'].isnull().sum()
```

```python
#Data type for all columns
df.dtypes
```

```python
df['budget'].mean()
df['gross'].mean()
```

```python
df['budget'].fillna(df['budget'].mean(), inplace = True)
df['gross'].fillna(df['gross'].mean(), inplace = True)
```

```python
#change data type of column
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df['rating'].__len__()
```

```python
#display max rows dataset
# pd.set_option('display.max_rows', None)
```

```python
df.sort_values(by = 'gross', ascending=False)
```

```python
#drop duplicate
df['company'] = df['company'].drop_duplicates()
```

```python
#Scatter plot budget with gross 
plt.scatter(x = df['gross'], y = df['budget'])
plt.title('Budget vs Gross')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for film')
plt.show()
```

```python
#plot budget vs gross using seaborn 
sns.regplot(x = df['gross'], y = df['budget'], line_kws={'color': 'red'})
```

```python
#Looking at correlation
df.corr(method = 'pearson')
```

```python
# high correlations between budget and gross
```

```python
correlation_matrix = df.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation matrix for numeric features')
plt.xlabel("Movie features")
plt.ylabel("Movie features")
```

```python
df_numerized = df.copy()
```

```python
 
for col in df_numerized.columns:
    if(df_numerized[col].dtype == 'object'):
         df_numerized[col] = df_numerized[col].astype('category')
         df_numerized[col] = df_numerized[col].cat.codes
```
